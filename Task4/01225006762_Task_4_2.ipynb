```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression


```

```python

data = pd.read_csv('shrink_ray_dataset.csv')
data.head()
```

```python

X = data[['Power']]
y = data['Shrinkage']


model = LinearRegression()
model.fit(X, y)
```

```python

shrink = 85
knob = (shrink - model.intercept_) / model.coef_
print('Predicted Knob for 85% Shrinkage:', knob[0])
```

```python

plt.scatter(data['Power'], data['Shrinkage'], label='Data')


x_range = np.linspace(0, 300, 100)
y_range = model.predict(x_range.reshape(-1, 1))
plt.plot(x_range, y_range, color='red', label='Linear Regression')

plt.xlabel('Knob')
plt.ylabel('Shrinkage')
plt.title('Plotting')
plt.legend()
plt.show()
```
